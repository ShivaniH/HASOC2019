```python
import pandas as pd
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("../outputs/DE/bert-base-multilingual-cased/task_4/train.tsv", sep="\t")
df.head()
```

```python
print(classification_report(df["label"], df["task_4_preds"]))
```

```python
for i in range(3):
    print(classification_report(
        df["label"].str.split("-", expand=True)[i], 
        df["task_4_preds"].str.split("-", expand=True)[i]
    ))
```
