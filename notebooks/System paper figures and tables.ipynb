```python
TASK_MAPPING={
    "task_1": "A",
    "task_2": "B",
    "task_3": "C",
    "task_4": "D",
}

MODEL_DETAILS = {
    ("EN", "A", "bert-base-cased"): ("1",),
    ("DE", "A", "bert-base-multilingual-cased"): ("1", ),
    ("HI", "A", "bert-base-multilingual-uncased"): ("1", ),
    ("EN", "B", "bert-base-cased"): ("1",),
    ("DE", "B", "bert-base-multilingual-cased"): ("1", ),
    ("HI", "B", "bert-base-multilingual-uncased"): ("1", ),
    ("EN", "C", "bert-base-cased"): ("1", ),
    ("HI", "C", "bert-base-multilingual-uncased"): ("1", ),
    
    ("HI", "A", "bert-base-multilingual-cased"): ("2", ),
    ("HI", "B", "bert-base-multilingual-cased"): ("2", ),
    ("HI", "C", "bert-base-multilingual-cased"): ("2", ),
    ("DE", "D", "bert-base-multilingual-cased"): ("2 (all)", ),
    ("EN", "D", "bert-base-uncased"): ("2 (all)",),
    
    ("HI", "D", "bert-base-multilingual-uncased"): ("3 (all)", ),
    ("DE", "D", "bert-base-german-cased"): ("3 (1)", ), # Only task 1
    ("DE", "B", "bert-base-german-cased"): ("3", ), # Only task 2
    ("EN", "A", "bert-base-uncased"): ("3", ),
    ("EN", "B", "bert-base-uncased"): ("3", ),
    ("EN", "C", "bert-base-uncased"): ("3", ),
}

TEST_RESULTS = {
    ("EN", "A", "bert-base-cased"): ("1", 0.7386, 0.7905, 14),
    ("EN", "B", "bert-base-cased"): ("1", 0.5175, 0.701, 3),
    ("EN", "C", "bert-base-cased"): ("1", 0.5002, 0.753, 2),
    
    ("EN", "A", "bert-base-uncased (D)"): ("2", 0.7465, 0.8012, 7),
    ("EN", "B", "bert-base-uncased (D)"): ("2", 0.537, 0.698, 2),
    ("EN", "C", "bert-base-uncased (D)"): ("2", 0.4765, 0.7639, 6),
    
    ("EN", "A", "bert-base-uncased"): ("3", 0.7399, 0.7903, 11),
    ("EN", "B", "bert-base-uncased"): ("3", 0.5446, 0.7277, 1),
    ("EN", "C", "bert-base-uncased"): ("3", 0.5111, 0.7563, 1),
    
    ("EN", "A", "best"): ("-", 0.7882, 0.8395, 1),
    ("EN", "B", "best (ours)"): ("-", 0.5446, 0.7277, 1),
    ("EN", "C", "best (ours)"): ("-", 0.5111, 0.7563, 1),
    
    ("HI", "A", "bert-base-multilingual-uncased"): ("1", 0.8018, 0.8025, 10),
    ("HI", "B", "bert-base-multilingual-uncased"): ("1", 0.5527, 0.6875, 7),
    ("HI", "C", "bert-base-multilingual-uncased"): ("1", 0.565, 0.7265, 2),
    
    ("HI", "A", "bert-base-multilingual-cased"): ("2", 0.8004, 0.801, 11),
    ("HI", "B", "bert-base-multilingual-cased"): ("2", 0.5534, 0.6755, 6),
    ("HI", "C", "bert-base-multilingual-cased"): ("2", 0.5492, 0.7484, 5),
    
    ("HI", "A", "bert-base-multilingual-uncased (D)"): ("3", 0.8108, 0.8141, 3),
    ("HI", "B", "bert-base-multilingual-uncased (D)"): ("3", 0.5812, 0.7147, 1),
    ("HI", "C", "bert-base-multilingual-uncased (D)"): ("3", 0.5503, 0.7583, 4),
    
    ("HI", "A", "best"): ("-", 0.8149, 0.8202, 1),
    ("HI", "B", "best"): ("-", 0.5812, 0.7147, 1),
    ("HI", "C", "best"): ("-", 0.5754, 0.7361, 1),
    
    
    ("DE", "A", "bert-base-multilingual-cased"): ("1", 0.5774, 0.7887, 4),
    ("DE", "B", "bert-base-multilingual-cased"): ("1", 0.2491, 0.7559, 12),
    
    #("DE", "A", "bert-base-multilingual-cased (D)"): ("2", np.nan, np.nan, -1), # Original from leaderboard
    ("DE", "A", "bert-base-multilingual-cased (D)"): ("2", 0.5734, 0.7986, 6), # Based on own computation
    ("DE", "B", "bert-base-multilingual-cased (D)"): ("2", 0.2758, 0.7779, 4),
    
    ("DE", "A", "bert-base-german-cased (D)"): ("3", 0.5216, 0.778, 12), # Only task 1    
    ("DE", "B", "bert-base-german-cased"): ("3", 0.2736, 0.7729, 6), # Only task 2
    
    ("DE", "A", "best"): ("-", 0.6162, 0.7915, 1),
    ("DE", "B", "best"): ("-", 0.3468, 0.7749, 1),
    
    
}


def get_model_D_stats(filepath, lang, model, task, split, run_id):
    print(filepath, lang, model, task, split, run_id)
    base_path = filepath.parent
    final_path = base_path / filepath.name.replace("_results.json", ".tsv")
    model = f"{model} (D)"
    run_id = run_id.split(" ") if run_id else []
    base_run_id = ""
    if len(run_id) == 2:
        base_run_id = run_id[0]
        run_id_tasks = run_id[1][1:-1]
        run_id_tasks = {1,2,3} if run_id_tasks == "all" else {int(run_id_tasks),}
    elif len(run_id) == 0:
        run_id_tasks = {}
    else:
        raise RuntimeError(f"Unknown run_id={run_id}")

    df = pd.read_csv(final_path, sep="\t")
    for i, task in enumerate(["A", "B", "C"]):
        run_id = ""
        if lang == "DE" and task == "C":
            continue
        if i+1 in run_id_tasks:
            run_id = base_run_id
        report = classification_report(
            df["label"].str.split("-", expand=True)[i], 
            df["task_4_preds"].str.split("-", expand=True)[i],
            output_dict=True
        )
        final_stat = {
            "macro avg/f1-score": report["macro avg"]["f1-score"], 
            "weighted avg/f1-score": report["weighted avg"]["f1-score"]
        }
        final_stat.update(dict(lang=lang, model=model, task=task, split=split, run_id=run_id))
        yield final_stat

def get_final_stats(filepath):
    lang, model, task, basename = filepath.parts[2:]
    stats = get_filestats(filepath)
    split = basename.split("_")[0]
    task = TASK_MAPPING[task]
    default_model_details = ("", )
    run_id, = MODEL_DETAILS.get((lang, task, model,), default_model_details)
    if task == "D":
        return [
            final_stat 
            for final_stat in get_model_D_stats(
                filepath, lang=lang, model=model, task=task, split=split, run_id=run_id
            )
        ]
    final_stat = stats[-1]
    final_stat.update(dict(lang=lang, model=model, task=task, split=split, run_id=run_id))
    return [final_stat]
```

```python
DATA_PATHS = {
    "EN": {
        "train": "../data/raw/training_data/english_dataset.tsv",
        "dev": "../data/raw/teaser_data/english_dataset.tsv",
        "test": "../data/raw/test_data/english_dataset.tsv",
    },
    "DE": {
        "train": "../data/raw/training_data/german_dataset.tsv",
        "dev": "../data/raw/teaser_data/german_dataset.tsv",
        "test": "../data/raw/test_data/german_dataset.tsv",
    },
    "HI": {
        "train": "../data/raw/training_data/hindi_dataset.tsv",
        "dev": "../data/raw/teaser_data/hindi_dataset.tsv",
        "test": "../data/raw/test_data/hindi_dataset.tsv",
    }
}
DATA_COLUMNS = ["row_id", "text", "task_A", "task_B", "task_C"]
NUM_LANGUAGES = len(DATA_PATHS)
```

```python
stats = []
for lang, lang_splits in DATA_PATHS.items():
    for split, path in lang_splits.items():
        df = pd.read_csv(path, sep="\t").fillna("NULL")
        df.columns = DATA_COLUMNS[:len(df.columns)]
        print(lang, split, path, df.shape)
        if split == "test":
            for task in ["A", "B", "C"]:
                if lang == "DE" and task == "C":
                    continue
                count = df.shape[0]
                stats.append((lang, split, task, count)) 
        else:
            task_cols = df.filter(regex=r'task_*', axis=1).columns
            for col in task_cols:
                task = col.split("_")[1]
                if lang == "DE" and task == "C":
                    continue
                count = df.loc[~df[col].isin({"NULL", "NONE"})].shape[0]
                stats.append((lang, split, task, count)) 
df_t = pd.DataFrame(stats, columns=["lang", "split", "task", "size"])
df_t = df_t.set_index(["task", "lang", "split"])["size"].sort_index().unstack([1, 2]).fillna(0).astype(int)
display(df_t)
print(df_t.to_latex(bold_rows=True, multirow=True))
```
